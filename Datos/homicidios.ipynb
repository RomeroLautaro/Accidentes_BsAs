VAMOS A JUNTAR LAS DOS TABLAS (HECHOS Y VICTIMAS)

```python
# Suponiendo que 'nuevo_hecho' y 'victimas' son tus DataFrames y 'ID' es la columna que utilizarás para unirlos
dataframe_combinado = pd.merge(nuevo_hecho, victimas, on='ID', how='left')
dataframe_combinado
```

```python
def verificar_tipo_datos_y_nulos(dataframe_combinado):
    """
    Esta función verifica los tipos de datos y la cantidad de valores nulos en un DataFrame.

    Parámetros:
        df (DataFrame): El DataFrame que se va a analizar.

    Retorna:
        DataFrame: Un DataFrame que contiene información sobre los tipos de datos y valores nulos en el DataFrame dado.
    """
    campos = []
    tipos_datos = []
    no_nulos_porcentaje = []
    nulos_porcentaje = []
    nulos = []

    for columna in dataframe_combinado.columns:
        campos.append(columna)
        tipos_datos.append([type(dataframe_combinado[columna].iloc[0])])
        no_nulos_porcentaje.append(dataframe_combinado[columna].count() / len(dataframe_combinado) * 100)
        nulos_porcentaje.append((1 - dataframe_combinado[columna].count() / len(dataframe_combinado)) * 100)
        nulos.append(len(dataframe_combinado) - dataframe_combinado[columna].count())

    resultado = pd.DataFrame({
        'nombre_campo': campos,
        'tipo_datos': tipos_datos,
        'no_nulos_%': no_nulos_porcentaje,
        'nulos_%': nulos_porcentaje,
        'nulos': nulos
    })

    return resultado

# Llamada a la función y mostrar el resultado
resultado = verificar_tipo_datos_y_nulos(dataframe_combinado)
print(resultado)
```
